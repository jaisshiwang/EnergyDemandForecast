```python
import pandas as pd
```

```python
import os
print(os.path.exists('../data/time_series.csv'))
```

```python
data_15_min = pd.read_excel('../data/time_series.xlsx', sheet_name='15min')
data_15_min.head()
```

```python
columns = []
for i in range(len(data_15_min.columns)):
    if data_15_min.columns[i] == 'Unnamed: 0' :
        columns.append('utc_timestamp')
    elif data_15_min.columns[i] == 'Unnamed: 1' :
        columns.append('cet_cest_timestamp')
    else :
        columns.append(str(data_15_min.iloc[0,i]) + '_' + str(data_15_min.iloc[1,i]))
```

```python
print(columns)
```

```python
data_15_min = data_15_min.loc[11:, :]
data_15_min.head()
```

```python
data_15_min.columns = columns
data_15_min = data_15_min.reset_index(drop=True)
data_15_min.head()
```

```python
data_15_min['utc_timestamp'] = pd.to_datetime(data_15_min['utc_timestamp'], format='%Y-%m-%d %H:%M:%S')
data_15_min.set_index('utc_timestamp', inplace=True)
```
